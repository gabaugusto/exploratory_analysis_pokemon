```python
#Always look a sample of the file
dataset.head()
```

```python
#Check is any values in the dataset is Null
dataset.isna().any()
```

```python
#replace the missing values in Height, Weight and capture_rate with mean values 
dataset['height_m'] = dataset['height_m'].fillna((dataset['height_m'].mean()))
dataset['weight_kg'] = dataset['weight_kg'].fillna((dataset['weight_kg'].mean()))
dataset['capture_rate'] = dataset['capture_rate'].fillna((dataset['capture_rate'].mean()))
```

```python
get_type_1 = pd.get_dummies(dataset['type1'], prefix="type1") 
get_type_2 = pd.get_dummies(dataset['type2'], prefix="type2") 
```

```python
#Removing not used columns
dataset.drop(dataset.columns[0:19], axis=1,inplace=True)
dataset.drop(['percentage_male', 'type1', 'type2', 'name', 'percentage_male'], axis=1, inplace=True)
```

```python
#Joining the created columns
dataset = pd.concat([dataset, get_type_1, get_type_2], axis=1, join='inner')
```

```python
#Look to a file sample with these changes
dataset[150:200]
```

```python
#Get the column that i want to find values. My target
target = dataset['is_legendary']
```

```python
#Get everything else but the is_legendary column
dataset = dataset.drop(['is_legendary'], axis = 1)
```

```python
#Finally using SKLearn to split the files
from sklearn.model_selection import train_test_split
```

```python
#spliting the file in four: 
x_train, x_test, y_train, y_test = train_test_split(dataset, target, test_size = 0.3)
```

```python
#Seeing the size of the new the new variables 
print(dataset.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#ExtraTreesClassifier is my classifier 
from sklearn.ensemble import ExtraTreesClassifier
```

```python
#Creating the model 
model_one = ExtraTreesClassifier()
```

```python
#Setting samples to training
model_one.fit(x_train, y_train)
```

```python
#Setting x_train to training and compare them to y_train
result = model_one.score(x_train, y_train)
```

```python
#See the Accuracy of the train
print("ACC: ", result)
```

```python
predict = model_one.predict(x_test)
```
